```python
import datetime
def dateLoop(start,end):
    datestart=datetime.datetime.strptime(start,'%Y%m%d')
    dateend=datetime.datetime.strptime(end,'%Y%m%d')
    date_list = [] 
    while datestart<dateend:
        datestart+=datetime.timedelta(days=1)
        date_list.append(datestart.strftime('%Y%m%d')) 
    return date_list
```

```python
def gethtmltext(url):
    result = requests.get(url)
    return result.text
```

```python
def dynamicUrl():
    dates=dateLoop('20200201','20200202')
    #print(dates)
    #print(type(dates))
    for date in dates:
        dynamicUrl='https://www.playsport.cc/livescore.php?aid=3&gamedate=%s&mode=11'%(date)
        time.sleep(2)
        text=gethtmltext(dynamicUrl)
        return text
```
